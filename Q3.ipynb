### q3 : Find and plot via pia chart the total number of patients of each disease (diagnosis) with respect to the gender(Female, Male)

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
```

```python
data=pd.read_csv('jsrt_metadata.csv')
data=data.dropna()
data
```

```python
#male
data_male = data.loc[data["gender"] == "Male"]
data_male.groupby("diagnosis")["gender"].count().plot(kind = "pie", figsize = (6, 6),autopct='%1.1f%%')
```

```python
#female
df_female= data.loc[data["gender"] == "Female"]
df_female.groupby("diagnosis")["gender"].count().plot(kind = "pie", figsize = (6,6),autopct='%0.1f%%')
```

```python
#all
count=data['gender'].value_counts()
print(count)
plt.pyplot.pie(data['gender'].value_counts(),shadow=True,labels=['Female','Male'],autopct='%1.1f%%'
              ,startangle=0)
```
